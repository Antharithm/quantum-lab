# 03 · Grover's search

Finding a marked item among N takes ~N tries classically, but only ~√N with Grover's algorithm. Here we search 4 states and amplify `11`.

```python
from quantumlab import circuits, run

qc = circuits.grover_2q()
print(qc.draw(output='text'))
```

```python
counts = run.run_circuit(qc, shots=2000, save=False).counts
counts
```

```python
top = max(counts, key=counts.get)
print('Most likely outcome:', top, '— Grover amplified the marked state.')
```

**How it works:** the *oracle* flips the phase of the answer, and the *diffuser* amplifies its probability via interference. Read `quantumlab/circuits.py` → `grover_2q` to see each step.
