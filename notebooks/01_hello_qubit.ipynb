# 01 · Hello, qubit

A qubit can be 0, 1, or a *superposition* of both. The Hadamard gate (`H`) makes an equal 50/50 mix — a quantum coin flip.

Measuring collapses it to a definite 0 or 1 at random.

```python
from qiskit import QuantumCircuit
from quantumlab import run, viz

qc = QuantumCircuit(1, 1)
qc.h(0)          # put the qubit into superposition
qc.measure(0, 0)
print(qc.draw(output='text'))
```

### Run it many times
Because the outcome is random, we run many *shots* and look at the distribution.

```python
result = run.run_circuit(qc, shots=2000, save=False)
result.counts
```

### Look at the statevector (before measuring)
The simulator can show the underlying amplitudes — the 'weights' of each outcome.

```python
import quantumlab.viz as viz
qc_nomeasure = QuantumCircuit(1)
qc_nomeasure.h(0)
viz.statevector_table(qc_nomeasure)
```
