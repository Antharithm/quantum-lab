# 02 · Entanglement (the Bell state)

Entangle two qubits so their results are *always* correlated. You'll only ever see `00` or `11` — never `01` or `10`.

```python
from qiskit import QuantumCircuit
from quantumlab import run

qc = QuantumCircuit(2, 2)
qc.h(0)         # superposition on q0
qc.cx(0, 1)     # entangle q0 -> q1
qc.measure([0,1],[0,1])
print(qc.draw(output='text'))
```

```python
run.run_circuit(qc, shots=2000, save=False).counts
```

Notice there are **no** `01` or `10` outcomes. Measuring one qubit instantly tells you the other — that's entanglement.
